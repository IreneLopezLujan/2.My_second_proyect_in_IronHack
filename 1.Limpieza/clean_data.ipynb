```python
#exportando csv
os.chdir("../data")
dataset=pd.read_csv("hrly_Irish_weather.csv", encoding = "ISO-8859-1")
dataset.head(10)
```

```python
#quitamos nulos
dataset.dropna()
#Filtramos solo por los restaurantes que hay en Dublin y obviamos el resto de paises que no nos interesan
D_county=dataset[dataset["county"]=='Dublin']
D_county.info()
```

```python
#filtramos el dataset a los ultimos dos años, ya que no nos interesa el tiempo de muchos años atras
D_county=D_county[(D_county['date'] > '2018-01-01')]
D_county.info()
```

```python
#chequear que columnas tienen algun nulo
r = D_county.isnull().any()
print(r)
```

```python
#pasar a formato int y string algunas columnas
D_county[['rain',"temp","msl"]] = D_county[['rain',"temp","msl"]].apply(pd.to_numeric,errors='coerce')
D_county["station"]=D_county["station"].astype(str)
#crear una columna index para poder dejar la columnaa date sin tocar
D_county["Index"]= D_county["date"]
```

```python
#pasar la columna index a indice
D_county.set_index(D_county.Index)
#renombrar dataset
data=D_county
data
```

```python
# eliminar muestras malas
data = data[data['rain'] > -500]

#Asignar el "día" a cada entrada de fecha
data['day'] = data['date'].apply(lambda x: x.date())

# Obtener la hora, el día y la hora de cada marca de tiempo en el conjunto de datos
data['time_of_day'] = data['date'].apply(lambda x: x.time())
data['day_of_week'] = data['date'].apply(lambda x: x.weekday())    
data['hour_of_day'] = data['time_of_day'].apply(lambda x: x.hour)
# Marcar el mes para cada entrada para que podamos ver los patrones mensuales
data['month'] = data['date'].apply(lambda x: x.month)
# crear una columna con los dias que trabajaremos (lunes a sabado)
data['working_day'] = (data['day_of_week'] >= 0) & (data['day_of_week'] <= 6)

```

```python
data['not_raining'] = data['rain'] = 0.0
data['raining'] = data['rain'] > 0.0

# Crearemos una columna con los dias que tendremos peligro de mojarnos, ya que es dia laboral y mañana
#o noche que es cuando vamos al trabajo.

clasificacion_dia=['Early Morning','Late Night']
for x in clasificacion_dia:   
    cluster=data['session']== x

data['get_wet_cycling'] = (data['working_day']) & (cluster)

#los dias de vacaciones entonces serán 
#data['get_wet_not_cycling'] = (data['working_day']) & (data['raining'])
```

```python
#seleccionar columnas que son validas 

data_clean=pd.DataFrame(data[['rain', 'temp','day', 'time_of_day', 'day_of_week',
       'hour_of_day', 'month', 'working_day', 'session', 'raining',
       'get_wet_cycling','date']])
data_clean.head(10)
```

```python
# Observa solo los días laborables y crea un conjunto de datos diarios de días laborables:
wet_cycling = data_clean.groupby('day')['get_wet_cycling'].any()
wet_cycling = pd.DataFrame(wet_cycling).reset_index()

wet_cycling
```

```python
wet_cycling.to_csv('wet_cycling_S_N.csv')
wet_cycling
```

```python
#agrupar por dia
rainy_days = data_clean.groupby(['day']).agg({
        "rain": [("rain_si_no", lambda x: (x > 0.0).any()),
                 ("rain_amount", "sum"),("rain_data", "max"),("get_wet_cycling","any")],
     
        })    

rainy_days
```

```python
#eliminar el index
rainy_days.reset_index(drop=False, inplace=True) 
#darle un nombre al index
rainy_days.rename(columns={"":"date"}, inplace=True) 
#eliminar index innecesarios
rainy_days.columns = rainy_days.columns.droplevel(level=0)
rainy_days
```

```python
#guardar tabla
rainy_days.to_csv('rainy_days.csv')
```

```python
# Agregar la cantidad de horas de lluvia por día al conjunto de datos rainy_days.
temp = data_clean.groupby(["day", "hour_of_day"])['raining'].any()
temp = temp.groupby(level=[0]).sum().reset_index()
#renombrar column
temp.rename(columns={'raining': 'hours_raining'}, inplace=True)
temp
```

```python
#mergear en rainy_days la tabla temp
rainy_days = rainy_days.merge(temp, left_on='date', right_on='day', how='left')
rainy_days.drop('day', axis=1, inplace=True)


rainy_days
#guardar csv
rainy_days.to_csv('rainy_days_hours')
```

# LIMPIANDO RESTAURANTES EN EUROPA-TRIPADVISOR

```python
#exportando csv
data_rest=pd.read_csv('TripAd_restaurants_europe.csv')
data_rest.columns
data_rest.City.unique()
```

```python
#filtrando solo por la ciudad de Dublin 
data_rest_D=data_rest[data_rest["City"]=='Dublin']
data_rest_D
```

```python
#observando el ratio de nulos por columnas
na_ratio = ((data_rest.isnull().sum() / len(data_rest_D))*100).sort_values(ascending = False)
print(na_ratio)
```

```python
#quitando nulos
list_D=data_rest['Cuisine Style'].dropna()
```

```python
#la columna que nos interesa es el tipo de cocina de los restaurantes en dublin y esta muy sucia
#la pasamos a lita
list_D=data_rest['Cuisine Style'].values.tolist()
```

```python
#es un string de listas, se quiere pasar a una lista de string
data_rest['Cuisine Style']= data_rest['Cuisine Style'].str.replace(r'[^a-zA-Z\s]+|X{2,}', ' ')
```

```python
lista=list(data_rest['Cuisine Style'])
lista
```

```python
#finalmente es lista de string
values = "".join(map(str, lista))
v_split=values.split()
#nos damos cuenta que los restaurantes se clasifican nosolo por nacionalidades sino por tipo de comida x ej.suhi
#Tambien se meten palabras aleatorias como Options, Friendly...
v_split
```

```python

#Vamos a quedarnos solo con las nacionalidades, para ello cogemos una lista de nacionalidades de internet 
#si estas nacionalidades estan en la lista , se deja el valor. De esta manera limpiamos datos extraños
list_v=list(v_split)
set_unicos=set(list_v)
nationalities=pd.read_csv("nationalities.csv", engine = "c")
list_n=list(nationalities["Nationality"])
set_nationalities=set(list_n)
finales=set_unicos.intersection(set_nationalities)
lista_elegi=[x for x in finales for e in v_split if x in e]
```

```python
#por ultimo contamos los restaurantes en dublin segun  nacionalidad 
counts = Counter(lista_elegi)
counter=dict(counts)
df_res = pd.DataFrame(counter, index=["Total"])
#y lo ordenamos
df_res=df_res.T.sort_values('Total',ascending=False)
df_res.head(10)
```

```python
#guardando csv
df_res.to_csv('nationalities_trip.csv')
```
